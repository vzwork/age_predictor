```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from tensorflow.keras.models import load_model
```

```python
reconstructed_model = load_model("./trained_model")
```

```python
im = Image.open('./test.jpg')
```

```python
im = im.resize((256, 256))
```

```python
image_array = np.array(im)
```

```python
image_array.shape
```

```python
plt.imshow(image_array)
```

```python
image_array = np.divide(image_array, 255)
```

```python
image_array[0][0]
```

```python
image_array = image_array.reshape(1, 256, 256, 3)
```

```python
age = reconstructed_model.predict(image_array)
```

```python
age
```
